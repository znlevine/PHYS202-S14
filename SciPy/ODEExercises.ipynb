(a) Use Euler's method (not odeint) to solve the following first-order differential equation with given initial condition y(0) = 0.

```python
def Euler(StepSize):
    xmax = 1.+StepSize # maximum time
    dx = StepSize # time step
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
 

    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        y[i+1] = y[i] + dx*(x[i]+2*y[i]) #Adds onto Y based on function
        
    return x,y

x1,y1=Euler(.25)
x2,y2=Euler(.02)
print y1[1/.25]
print y2[1/.02]
```

(c) Now solve the same problem using Euler-Cromer and the Midpoint method for the same pair of step sizes and plot the results along with the exact solution. Then also plot the difference between the approximate numerical solutions and the exact solution. Which combination of method and step-size gives the best result?

```python
def EulerCromer(StepSize):
    xmax = 1.+StepSize # maximum time
    dx = StepSize # time step
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
 

    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        y[i+1] = y[i] + dx*(x[i+1]+2*y[i]) #Adds onto Y based on function
        
    return x,y

x1,y1=EulerCromer(.25)
x2,y2=EulerCromer(.02)
print y1[1/.25]
print y2[1/.02]

plt.plot(x1,y1)
plt.plot(x2,y2)
x=np.arange(0,1.01,.01)
y=.25*exp(2*x)-0.5*x-0.25
plt.plot(x,y)
plt.show()

def Midpoint(StepSize):
    xmax = 1.+StepSize # maximum time
    dx = StepSize # time step
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
 

    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        y[i+1] = y[i] + dx*((x[i+1]+x[i])/2+2*y[i]) #Adds onto Y based on function
        
    return x,y

x1,y1=Midpoint(.25)
x2,y2=Midpoint(.02)
print y1[1/.25]
print y2[1/.02]

plt.plot(x1,y1)
plt.plot(x2,y2)
x=np.arange(0,1.01,.01)
y=.25*exp(2*x)-0.5*x-0.25
plt.plot(x,y)
plt.show()

```

```python
#Angle of 50, speed of 30 m/s, 0-200 sec, x0,y0=0,0, vter=44m/s

theta=50./360.*(2*np.pi)
v=30.
g=9.8
x0=0.
y0=0.
vter=44.
tmax=200.01
tpoints=np.arange(0,tmax,0.01)
vx=(v*np.cos(theta))
vy=(v*np.sin(theta))#add into function to recalculate

derivx=lambda velx,vely: -g*velx/(vter**2)*np.sqrt(velx**2+vy**2)
derivy=lambda velx,vely: -g*(1+vely/(vter**2)*np.sqrt(velx**2+vely**2))

xderived=odeint(derivx,vx,tpoints)
yderived=odeint(derivy,vy,tpoints)
print xderived
print yderived
plt.subplot(211)
plt.plot(tpoints,xderived)
plt.subplot(212)
plt.plot(tpoints,yderived)
plt.ylim(0,20)
plt.show()
```
