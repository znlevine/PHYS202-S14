# Numerical Integration

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
func = lambda x: x**4 - 2*x + 1
```

```python
N=10
a=0.0
b=2.0
h=(b-a)/N

k=np.arange(1,N)
I=h*(0.5*func(a)+0.5*func(b)+func(a+k*h).sum())

print "Trapezoidal Rule Integral =",I
```

```python
N=10
a=0.0
b=2.0
h=(b-a)/N

k1=np.arange(1,N/2+1)
k2=np.arange(1,N/2)
I=(1./3.)*h*(func(a)+func(b)+4.*func(a+(2*k1-1)*h).sum()+2.*func(a+2*k2*h).sum())

print "Simpson's Rule Integral=",I
```

```python
x=np.arange(0,20,2)
y=np.array([0,3,5,2,8,9,0,-3,4,9],dtype=float)
plt.plot(x,y)

plt.fill_between(x,y,y2=0,color='red',hatch='//',alpha=0.2)

plt.show
```

```python
I=integrate.simps(y,x)
print "Simpson's Rule Integral for Discrete Samples =",I
```

```python
from scipy.integrate import dblquad

#IMPORTANT! Y must come first in integrand!!!
integrand=lambda y,x: y*np.sin(x)+x*np.cos(y)

xmin=np.pi
xmax=2*np.pi

#Callable functions for y limits are contsants:
ymin = lambda x:0
ymax = lambda x:np.pi

#See help for argument order
ans,err=dblquad(integrand,xmin,xmax,ymin,ymax)
print "The double integral is ",ans
```

```python
from scipy.integrate import tplquad

#Inner to outer - start with Z, then y, then x

integrand = lambda z,y,x: y*np.sin(x)+z*np.cos(x)
xmin=0
xmax=np.pi

ymin=lambda x:0
ymax=lambda x:1

#Note order of these arguments
zmin = lambda x,y:-1
zmax = lambda x,y:1

#Order of limits is outer to inner
ans,err=tplquad(integrand,xmin,xmax,ymin,ymax,zmin,zmax)
print "The triple integral = ",ans
```
